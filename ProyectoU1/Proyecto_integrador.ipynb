**1. Variables X y Y**

La variable independiente (X) son las palabras leidas, porque es la que se utiliza para explicar o predecir el resultado. La variable dependiente (Y) es el tiempo en segundos, ya que depende de las palabras para medir el tiempo que tardan leyendo.

```python
import pandas as pd
df = pd.DataFrame({
  "tiempo_en_segundos":(32.38,46.75,31.44,32.29,29.93,85.21,56.18,31.79,89.09,35.07,
   54.6,42.15,35.83,46.97,31.98),
  "palabras_leidas":(142,128,171,181,100,274,192,149,177,167,186,123,80,126,79)}) # Missing parenthesis added here

Y = df['tiempo_en_segundos']
X = df['palabras_leidas']
```

**2. Diagrama de dispersión de datos**

```python
# Importamos la librería matplotlib.pyplot y le damos el pseudónimo plt
import matplotlib.pyplot as plt

# Configuración general del gráfico
plt.figure(
    figsize=(6, 4),   # tamaño de la figura (ancho, alto) en pulgadas
    dpi=120           # resolución del gráfico
)

# Gráfico de dispersión
plt.scatter(
    X, Y,
    marker="*",       # forma: googlear "matplotlib.markers"
    color='purple',     # color de los puntos
    edgecolor='purple',    # borde de los puntos
    alpha=0.9,            # transparencia: 0 es invisible y 1 es color sólido
    s=50,                 # tamaño de los puntos
)

# Etiquetas de los ejes
# eje x
plt.xlabel(
    'palabras_leidas', # etiqueta del eje x
    fontsize=12  # tamaño de fuente
)

plt.ylabel(
    'tiempo_en_segundos', # etiqueta del eje y
    fontsize=12 # tamaño de fuente
)

# Fuente de los ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Márgenes
plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior

# Cuadrícula (opcional, pero didáctica)
plt.grid(
    visible=True,
    linestyle=':',
    linewidth=0.7,
    alpha=0.5,
    color="pink"
)
```

El diagrama de dispersión muestra una tendencia positiva entre las variables, conforme aumenta el número de palabras leídas, también aumenta el tiempo empleado en la lectura. Aunque existe cierta dispersión entre los puntos, se observa una relación ascendente.

**3. ¿Los datos soportan la suposición de linealidad?**

Sí, los datos soportan la suposición de linealidad, ya que los puntos presentan una tendencia positiva y siguen aproximadamente una línea recta. Aunque existe cierta variabilidad, el comportamiento general permite ajustar un modelo de regresión lineal.

**4. Coeficiente de correlación**

El coeficiente de correlación es $r = 0.6428$, lo que indica una correlación positiva moderada entre las palabras leídas y el tiempo en segundos, ya que conforme aumenta el número de palabras leídas, también aumenta el tiempo requerido para leerlas.

```python
from scipy.stats import pearsonr

# Define X and Y from the DataFrame to ensure they are available
Y = df['tiempo_en_segundos']
X = df['palabras_leidas']

# Test de Pearson
# H0: rho = 0     (No hay correlación)
# H1: rho ≠ 0     (Sí hay correlación)
# alpha = 0.05

r, valor_p = pearsonr(X, Y)

print(f"Coeficiente de correlación: {r: 0.4f}")
print(f'valor_p: {valor_p: 0.4f}')
```

5. **Calcule el coeficiente de determinación e interprete el resultado.**

```python
import pandas as pd
import statsmodels.api as sm

df = pd.DataFrame({
  "tiempo_en_segundos":(32.38,46.75,31.44,32.29,29.93,85.21,56.18,31.79,89.09,35.07,
   54.6,42.15,35.83,46.97,31.98),
  "palabras_leidas":(142,128,171,181,100,274,192,149,177,167,186,123,80,126,79)})

Y = df['tiempo_en_segundos']
X = df['palabras_leidas']

x_constante = sm.add_constant(X)
modelo = sm.OLS(Y, x_constante).fit()
y_calculada = modelo.predict(x_constante)
```

```python
# Coeficiente de determinación
from sklearn.metrics import r2_score

r2 = r2_score(Y, y_calculada)
print(f"Coeficiente de determinación: {r2: 0.2%}")
```

El coeficiente de determinación es $R² = 41.32%$, lo que significa que el $41.32%$ de la variación del tiempo de lectura puede explicarse por el número de palabras leídas, mientras que el $58.68%$ restante se debe a otros factores que no están considerados en el modelo.

12. **Realice una tabla ANOVA e interprete el resultado.**

```python
# test de ANOVA (Analysis of Variance)
# H0: beta_1 = 0   (No hay correlación)
# H1: beta_1 ≠ 0   (Sí hay correlación)

import statsmodels.api as sm
from statsmodels.formula.api import ols
# Y ~ X
modelo_lineal = ols(
    '''tiempo_en_segundos ~ palabras_leidas''',
    data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
```

Se obtuvo un valor F = $9.1543$ y un valor-p = $0.0097$, el cual es menor que 0.05. Por ello, se rechaza la hipótesis nula y se concluye que existe una relación lineal significativa entre el número de palabras leídas y el tiempo en segundos, por lo que el modelo de regresión es adecuado para explicar los datos.
