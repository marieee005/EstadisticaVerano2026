```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/marieee005/EstadisticaVerano2026/refs/heads/main/GymMembers/gym_members_exercise_tracking.csv")

df
```

**1. Variables X y Y**


+ $Y$: Calories Burned: Es la variable dependiente ya que se busca saber si aumenta el tiempo de entrenamiento tambien aumenta las calorías quemadas.
+ $X$: Session Duration (hours): Es porque es la variable independiente, puede influir en el resultado.

```python
Y = df["Calories_Burned"] # variable dependiente
X = df["Session_Duration_(hours)"] # variable independiente
```

**2. Diagrama de dispersión de datos**

```python
# Marielena Carvajala Becerra
# Importamos la librería matplotlib.pyplot y le damos el pseudónimo plt
import matplotlib.pyplot as plt

# Configuración general del gráfico
plt.figure(
    figsize=(6, 4),   # tamaño de la figura (ancho, alto) en pulgadas
    dpi=120           # resolución del gráfico
)

# Gráfico de dispersión
plt.scatter(
    X, Y,
    marker="o",       # forma: googlear "matplotlib.markers"
    color='#27D6F5',     # color de los puntos
    edgecolor='black',    # borde de los puntos
    alpha=0.9,            # transparencia: 0 es invisible y 1 es color sólido
    s=50,                 # tamaño de los puntos
)

# Etiquetas de los ejes
# eje x
plt.xlabel(
    'Duración en horas', # etiqueta del eje x
    fontsize=12  # tamaño de fuente
)

plt.ylabel(
    'Calorías quemadas', # etiqueta del eje y
    fontsize=12 # tamaño de fuente
)

# Fuente de los ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Márgenes
plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior

# Cuadrícula (opcional, pero didáctica)
plt.grid(
    visible=True,
    linestyle='--',
    linewidth=0.7,
    alpha=0.5,
    color="gray"
)
```

En la gráfica se observa que, cuando aumenta la duración del entrenamiento, también aumentan las calorías quemadas. Esto muestra una relación positiva entre las variables.


**3. ¿Los datos soportan la suposición de linealidad?**

Los datos sí apoyan la idea de una relación lineal, porque los puntos siguen una tendencia ascendente bastante clara. Es decir, entre más tiempo dura la sesión, más calorías se queman.

**4. Coeficiente de correlación**

```python
# Importar la función pearsonr desde scipy.stats
from scipy.stats import pearsonr

# Test de Pearson
# H0: rho = 0     (No hay correlación)
# H1: rho ≠ 0     (Sí hay correlación)
# alpha = 0.05

r, valor_p = pearsonr(X, Y)

print(f"Coeficiente de correlaciòn: {r: 0.4f}")
print(f'valor_p: {valor_p: 0.4f}')
```

El coeficiente de correlación es 0.9081, lo que indica una relación positiva muy fuerte. Esto significa que la duración del ejercicio y las calorías quemadas están muy relacionadas.

**5. Coeficiente de Determinación**

```python
import statsmodels.api as sm
x_constante = sm.add_constant(X)
modelo = sm.OLS(Y, x_constante).fit()
y_calculada = modelo.predict(x_constante)
```

```python
# Coeficiente de determinación
from sklearn.metrics import r2_score

r2 = r2_score(Y, y_calculada)
print(f"Coeficiente de determinación: {r2: 0.2%}")
```

El coeficiente de determinación es 82.47%. Esto significa que la duración del entrenamiento explica aproximadamente el 82.47% de los cambios en las calorías quemadas.

+ **Independencia:** No se observa un patrón claro entre los residuales, por lo que se considera que son independientes.

**9. Prueba de Shapiro**

**10. Prueba de Breusch-Pagan**

```python
# Test de Breusch-Pagan
# H0: No hay heterocedasticidad (dispersión es igual en toda la recta)
# H1: Hay Heterocedasticidad (dispersión distinta a lo largo de la recta)
# alpha = 0.05

from statsmodels.stats.api import het_breuschpagan
import statsmodels.api as sm
_, valor_p_bp, _, _ = het_breuschpagan(residuales, x_constante)
print(f'valor-p de Breusch-Pagan: {valor_p_bp: 0.4f}\n')
```

El valor-p es 0.0000, menor que 0.05. Esto significa que no se cumple la homoscedasticidad, es decir, la dispersión de los errores no es igual en todos los valores.

**12. Tabla ANOVA**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Clean column names to remove special characters and spaces for the formula
df_cleaned = df.copy()
df_cleaned.rename(columns={'Calories_Burned': 'Calories_Burned_', 'Session_Duration_(hours)': 'Session_Duration_hours'}, inplace=True)

# Y ~ X
modelo_lineal = ols('Calories_Burned_ ~ Session_Duration_hours', data=df_cleaned).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
```

La tabla ANOVA muestra un valor-p de 0.0000, lo que indica que el modelo de regresión es significativo. Esto quiere decir que la duración de la sesión sí ayuda a explicar las calorías quemadas.
